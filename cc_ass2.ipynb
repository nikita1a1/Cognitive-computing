```python
import random
list1=list(range(101,900))
print(list1)
list2=list(random.choices(list1,k=100))
list3=[]
list4=[]
list5=[]
print(list2)
count1=0
count2=0
count3=0
for i in list2:
  if(i%2==0):
    count1=count1+1
    list3.append(i)

  else:
    count2=count2+1
    list4.append(i)
for i in list2:
  co=False
  for k in range (2,int(i/2)+1):
   if(i%k==0):
    co=True
    break
  if(co==False):
   count3=count3+1
   list5.append(i)


print("prime numbers are ",count3)
print(list5)
print("even numbers are ",count1)
print(list3)
print("odd numbers are ",count2)
print(list4)


```
